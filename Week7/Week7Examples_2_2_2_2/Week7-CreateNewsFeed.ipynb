```python
#  Create our own news feed  ##
#  MSN RSS as an example ##
import urllib.request
import xml.etree.ElementTree as ET
import io
url = "https://rss.msn.com/"
xmlstring = urllib.request.urlopen(url).read().decode('utf8')  #Convert MSN feed into string
#print(len(xmlstring)) #Determine the number of characters
root = ET.fromstring(xmlstring) #Create a tree with root and child starting from Channel
for child in root.iter('item'): #Iterate throught all items
     title = child.find('title').text #Note title format
     link = child.find('link').text  #Note link in text format
     print("NewsFlash:",title,"\n", "Full Story:",link,"\n\n")   #Create news feed
```
