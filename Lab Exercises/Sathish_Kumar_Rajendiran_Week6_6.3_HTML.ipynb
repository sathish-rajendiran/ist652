```python
from urllib import request
```

```python
snapchaturl = 'http://www.bbc.com/news/business-39135278'
```

```python
response = request.urlopen(snapchaturl).read().decode('utf8')
```

```python
type(response)
print(response[:500])
```

```python
from bs4 import BeautifulSoup
```

```python
responsesoup = BeautifulSoup(response, 'html.parser')
responsesoup
```

```python
firsttitle = responsesoup.title
firsttitle
```

```python
print(type(firsttitle))
print(firsttitle.name)
print(firsttitle.get_text())
```

```python
alltitles = responsesoup.find_all('title')
len(alltitles)
alltitles
```

```python
anchors = responsesoup.find_all('a')
len(anchors)
print(anchors[:8])
```

```python
links = [str(link.get('href')) for link in responsesoup.find_all('a')]
print(links[ :4])
```

```python
outlinks = [link for link in links if link.startswith('http')]
len(outlinks)
print(outlinks[:4])
```

```python
from urllib.request import Request, urlopen
from urllib.error import URLError

espnurl = 'https://www.espn.com/nfl/standings22'
try:
    resp = request.urlopen(espnurl).read().decode('utf8')
except URLError as e:
    if hasattr(e, 'reason'):
        print('We failed to reach a server.')
        print('Reason: ', e.reason)
    elif hasattr(e, 'code'):
        print('The server couldn\'t fulfill the request.')
        print('Error code: ', e.code)
else:
    respsoup = BeautifulSoup(resp,'lxml')
    print(resp[:500])
    
```

```python
rtitle = respsoup.title
rtitle.get_text()
```
